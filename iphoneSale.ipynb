```python
# Define price bins and labels
bins = [0, 10000, 30000, 50000,70000,90000, data['Sale Price'].max()]
labels = ['<10000', '10000-30000', '30001-50000', '50001-70000','70001-90000','>90000']

# Create a new column for price range
data['Price Range'] = pd.cut(data['Sale Price'], bins=bins, labels=labels)
print (data['Price Range'])

# Count products in each price range
price_range_counts = data['Price Range'].value_counts().sort_index()
print(price_range_counts )

# Plot
plt.figure(figsize=(20,10))
plt.bar(price_range_counts.index, price_range_counts.values, color='blue', edgecolor='black')
plt.xlabel('Sale Price Range')
plt.ylabel('Number of Products')
plt.title('Number of iPhones in Each Sale Price Range')
plt.show()
```

Now, The relationship between the sale price of iPhones and their ratings on Flipkart:

```python
# We'll use count of ratings as x-axis data
count_of_rating= data['Number Of Ratings']
count_of_rating
# And, sale price as y-axis data
data['Sale Price']

print(count_of_rating,data['Sale Price'])
```

```python
# We use polyfit() function to use trendlines
x = np.polyfit(count_of_rating,data['Sale Price'],1)
# polyfit() takes 3 arguments: x,y, degree of polynomial
# x=x-axis=count_of_ratings2; y=y-axis=data['Sale Price'] and degree=1 for a straight line

x1 = np.poly1d(x)#creates a polynomial function (x1) you can use to compute the trendline values for plotting
```

```python
plt.figure(figsize=(20,10))
plt.scatter(count_of_rating,data['Sale Price'],marker='x',alpha=0.5)
# Trend Lines do provide a better visual to view the relationships observed via scatter plots
x = np.polyfit(count_of_rating,data['Sale Price'],1)
x1 = np.poly1d(x)
plt.plot(count_of_rating,x1(count_of_rating),'o-')

plt.xlabel('Count Of Ratings')
plt.ylabel('Sale Price')
plt.title('Relationship b/w Sale Price and Number of ratings')

plt.show()
```

There is a negative linear relationship between the sale price of iPhones and the number of ratings

This means that iPhones with lower sale prices are sold more in India



Now,The relationship between the discounts offered for iPhones and their ratings on Flipkart:

```python


print(count_of_rating,data['Discount Percentage'])
```

```python
# We use polyfit() function to use trendlines
z = np.polyfit(count_of_rating,data['Discount Percentage'],1)

z1 = np.poly1d(z)
plt.figure(figsize=(20,10))
plt.scatter(count_of_rating,data['Discount Percentage'],marker='x',alpha=0.5)
# Trend Lines do provide a better visual to view the relationships observed via scatter plots
z = np.polyfit(count_of_rating,data['Discount Percentage'],1)
z1 = np.poly1d(z)
plt.plot(count_of_rating,z1(count_of_rating),'o-')

plt.xlabel('Count Of Ratings')
plt.ylabel('Discount Percentage')
plt.title('Relationship b/w Discount Percentages and Number of ratings')

plt.show()
```

As observed from the above figure, there is a positive linear relationship now between the number of ratings when the discounts come into picture

This means that for iPhones to sell effectively, suitable discounts need to be offered in order to garner more product sales

Hence, irrespective of the hardware and phone specs, discounts offered probably plays a primary role in the product sales in India
